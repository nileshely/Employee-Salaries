# EMPLOYEE SALARIES ANALYSIS REPORT

The dataset comprises annual salary information for active, permanent employees of Montgomery County, MD, for the calendar year 2023. It includes data on gross pay, overtime pay, department, division, gender, base salary, and grade. This dataset presents an opportunity for comprehensive analysis to understand salary disparities, departmental distributions, and the impact of factors like overtime and longevity pay on overall compensation. Exploring this data could unveil patterns and insights crucial for ensuring fairness and equity within the organization's compensation structure.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/kaggle/input/employee-salaries-analysis/Employee_Salaries.csv")
```

```python
data.head()
```

```python
data.describe()
```

Gender Pay Gap Analysis reveals notable differences between genders within similar roles, grades, and departments. By examining the distribution of salaries across genders, we can identify disparities that require attention and rectification. Departmental Salary Analysis demonstrates variations in compensation across different departments and divisions, shedding light on how salary structures differ within the organization. Additionally, evaluating the Impact of Overtime and Longevity Pay unveils trends in these payments and their contribution to overall compensation. By employing statistical models and machine learning techniques, predictive analyses can offer insights into base salaries based on factors such as department, grade, and length of service.

```python
# Set purple color shades
purple = ['#9e9ac8', '#807dba', '#6a51a3', '#54278f', '#3f007d']
```

```python
# Visualization 1: Gender distribution
plt.figure(figsize=(8, 6))
data['Gender'].value_counts().plot(kind='bar', color=purple)
plt.title('Gender Distribution of Employees')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Department-wise salary distribution
plt.figure(figsize=(10, 14))
data.groupby('Department_Name')['Base_Salary'].mean().sort_values().plot(kind='barh', color=purple)
plt.title('Average Salary by Department')
plt.xlabel('Average Salary')
plt.ylabel('Department')
plt.show()
```

```python
# Visualization 3: Distribution of overtime pay
plt.figure(figsize=(8, 6))
plt.hist(data['Overtime_Pay'], bins=30, color=purple[2])
plt.title('Distribution of Overtime Pay')
plt.xlabel('Overtime Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 4: Distribution of longevity pay
plt.figure(figsize=(8, 6))
plt.hist(data['Longevity_Pay'], bins=30, color=purple[3])
plt.title('Distribution of Longevity Pay')
plt.xlabel('Longevity Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 5: Scatter plot of base salary vs. overtime pay
plt.figure(figsize=(8, 6))
plt.scatter(data['Base_Salary'], data['Overtime_Pay'], color=purple[4], alpha=0.5)
plt.title('Base Salary vs. Overtime Pay')
plt.xlabel('Base Salary')
plt.ylabel('Overtime Pay')
plt.show()
```

```python
# Visualization 6: Box plot of base salary by gender
plt.figure(figsize=(8, 6))
data.boxplot(column='Base_Salary', by='Gender', grid=False, color=dict(boxes=purple[2], whiskers=purple[1], medians=purple[4], caps=purple[3]))
plt.title('Base Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Base Salary')
plt.suptitle('')
plt.show()
```

```python
# Visualization 7: Pie chart of gender distribution
plt.figure(figsize=(8, 6))
data['Gender'].value_counts().plot(kind='pie', autopct='%1.2f%%', colors=purple)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Visualization 8: Pairwise scatter plot of numerical variables
sns.pairplot(data[['Base_Salary', 'Overtime_Pay', 'Longevity_Pay']], diag_kind='kde', plot_kws={'color': purple[2]})
plt.show()
```

In conclusion, this dataset provides valuable insights into employee compensation within Montgomery County, MD. Through thorough analysis, we've uncovered disparities in gender pay, variations in departmental salary distributions, and trends in overtime and longevity pay. Addressing these disparities and understanding the factors influencing compensation is crucial for fostering a fair and equitable work environment. Moving forward, implementing measures to mitigate gender pay gaps, ensuring transparency in salary structures, and continuously evaluating compensation practices are essential for promoting fairness and equity within the organization.
